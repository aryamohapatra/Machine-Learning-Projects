```python
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from keras.models import Model, load_model
from keras.layers import Input, Dense
from keras.callbacks import ModelCheckpoint, TensorBoard
from keras.callbacks import EarlyStopping ,ReduceLROnPlateau
from keras.optimizers import Adam
```

```python
df_class = df['Class']
df_features = df.drop('Class',axis=1)
df_features = df_features.drop('Time',axis=1)
minmax_scaler = preprocessing.MinMaxScaler()
X = minmax_scaler.fit_transform(df_features.values)
df_normalized = pd.DataFrame(X)
```

```python
df_normalized['Class'] = df['Class']
```

```python
normal_tran = df_normalized[df_normalized['Class']==0]
fraud_tran = df_normalized[df_normalized['Class']==1]
normal_tran_test_1=normal_tran.sample(frac=0.05)
normal_tran=normal_tran.drop(normal_tran_test_1.index)
normal_tran_test_2=normal_tran.sample(frac=0.05)
normal_tran=normal_tran.drop(normal_tran_test_2.index)
```

```python
normal_tran.shape
```

```python
fraud_tran_test = fraud_tran.sample(frac=0.05)
fraud_tran_val = fraud_tran.drop(fraud_tran_test.index)
tran_test=normal_tran_test_1.append(fraud_tran_test)
tran_val=normal_tran_test_2.append(fraud_tran_val)
```

```python
print("train set dimensions :",normal_tran.shape)
print("test set dimensions :",tran_test.shape)
print("validate set dimensions :",tran_val.shape)
```

```python
X_train, X_test= train_test_split(normal_tran,test_size=0.2, random_state=2020)
```

```python
X_train = X_train.drop(['Class'],axis=1).values
y_test = X_test['Class']
X_test = X_test.drop(['Class'],axis=1).values
```

```python
input_dim = X_train.shape[1]
encoding_dim = 20
input_layer = Input(shape=(input_dim, ))
encoder = Dense(encoding_dim*2, activation="relu")(input_layer)
encoder = Dense(encoding_dim, activation="relu")(encoder)
encoder = Dense(8,activation="relu")(encoder)
decoder = Dense(20, activation='relu')(encoder)
decoder = Dense(40, activation='relu')(encoder)
decoder = Dense(input_dim, activation='sigmoid')(decoder)
autoencoder = Model(inputs=input_layer, outputs=decoder)

nb_epoch = 50
batch_size = 32
autoencoder.compile(optimizer='adam', 
                    loss='mean_squared_error', 
                    metrics=['accuracy'])
es = EarlyStopping(monitor='val_loss', mode='min', verbose=1,patience=15)

checkpointer = ModelCheckpoint(filepath="model.h5",
                               verbose=0,
                               save_best_only=True)
```

```python
import tensorflow as tf
autoencoder = tf.keras.models.load_model('model.h5')
```

```python
autoencoder.summary()
```

```python
history = autoencoder.fit(X_train, X_train,
                    epochs=nb_epoch,
                    batch_size=batch_size,
                    shuffle=True,
                    validation_data=(X_test, X_test),callbacks=[es,checkpointer],
                    verbose=1)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper right')
plt.show()
```

```python
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.title('model acc')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='lower right')
plt.show()
```

```python
import numpy as np
predictions = autoencoder.predict(X_test)
mse = np.mean(np.power(X_test - predictions, 2), axis=1)
error_df = pd.DataFrame({'reconstruction_error': mse,
                        'true_class': y_test})
error_df.describe()
```

```python
error_df_val["predicted_class"]=[1 if x > 0.001 else 0 for x in error_df_val["reconstruction_error"]]
```

```python
from sklearn.metrics import (confusion_matrix, precision_recall_curve, auc,
                             roc_curve, recall_score, classification_report, f1_score,
                             precision_recall_fscore_support)

print(classification_report(error_df_val["true_class"],error_df_val["predicted_class"]))
```

```python
import matplotlib.pyplot as plt
con_mat = confusion_matrix(error_df_val["true_class"], error_df_val["predicted_class"]) 
df_con_mat = pd.DataFrame(con_mat, index = [i for i in ('Normal','Fault')], columns = [i for i in ('True','False')]) 
plt.figure(figsize = (8,4)) 
sns.heatmap(df_con_mat, annot=True,fmt="d") 
plt.title('Confusion Matrix') 
plt.xlabel("Predicted") 
plt.ylabel("Actuals")
```

```python
fpr, tpr, thresholds = roc_curve(error_df_val.true_class, error_df_val.reconstruction_error)
roc_auc = auc(fpr, tpr)

plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr, label='AUC = %0.4f'% roc_auc)
plt.legend(loc='lower right')
plt.plot([0,1],[0,1],'r--')
plt.xlim([-0.001, 1])
plt.ylim([0, 1.001])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show();
```
